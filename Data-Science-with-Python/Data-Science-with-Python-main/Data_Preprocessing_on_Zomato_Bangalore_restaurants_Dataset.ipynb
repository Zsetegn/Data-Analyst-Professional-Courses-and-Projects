```python
import pandas as pd
```

```python
zomato = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/zomato.csv",engine="python",encoding='cp1252')
```

```python
#plot the count of rating.
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = 14,7
sns.countplot(zomato["rate"], palette="Set1")
plt.title("Count plot of rate variable")
plt.show()
```

image.png

```python
#joint plot for 'rate' and 'votes'
sns.jointplot(x = "rate", y = "votes", data = zomato, height=8, ratio=4, color="g")
plt.show()
```

```python
#analyze the number of restaurants in a location
zomato.location.value_counts().nlargest(10).plot(kind = "barh")
plt.title("Number of restaurants by location")
plt.xlabel("Count")
plt.show()
```

image.png

```python
#restaurants serve to
zomato.serve_to.value_counts().nlargest(10).plot(kind = "barh")
plt.title("Number of restaurants listed in a particular location")
plt.xlabel("Count")
plt.show()
```

image.png

```python
sns.countplot(hue = zomato["online_order"], palette = "Set1", x = zomato["rate"])
plt.title("Distribution of restaurant rating over online order facility")
plt.show()
```
