## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python

QUERY = """
SELECT DatabaseWorkedWith,COUNT(*) as count
FROM DatabaseWorkedWith
group by DatabaseWorkedWith
order by count desc
"""
df5 = pd.read_sql_query(QUERY,conn).head()
df6 = pd.read_sql_query(QUERY,conn)
df5

```

```python
plt.figure(figsize=(6,6))
plt.pie(df5['count'], labels = df5.DatabaseWorkedWith, startangle=90, autopct='%.1f%%')
plt.title('Top five Databse respondents for next year')
plt.show()
```

```python
import plotly.graph_objects as go

labels = df5['DatabaseWorkedWith']
values = df5['count']

fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
import plotly.express as px
fig = px.pie(df5, values='count', names='DatabaseWorkedWith')
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')
fig.show()
```

```python
import plotly.express as px
fig = px.pie(df5, values='count', names='DatabaseWorkedWith',
             title='Top five Databse respondents for next year',
             hover_data=['count'], labels={'count':'Db count'})
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs ,CodeRevHrs
FROM master
where Age >=30 AND Age <=35
"""
df7 = pd.read_sql_query(QUERY,conn)
df8 = df7.groupby('Age').median()

df8.plot(kind='bar', figsize=(10, 6), stacked=True)

plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs for Those Age 30 to 35')
plt.show()


```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT Age, ConvertedComp
FROM master
where Age BETWEEN 45 AND 60
"""
df9 = pd.read_sql_query(QUERY,conn)
n_by_state = df9.groupby("Age")["ConvertedComp"].median()

# importing the required libraries
import matplotlib.pyplot as plt
import numpy as np

plt.plot(n_by_state)  # Plot the chart
plt.xlabel("Age")
plt.ylabel("median ConvertedComp")
plt.title('Line Chart of Median ConvertedComp  for Those Age 45 to 60')
plt.show()  # display
```
