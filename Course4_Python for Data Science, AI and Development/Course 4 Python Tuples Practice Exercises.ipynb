```python
#Q19 check whether Coursera exists within  myTuple

# NOTE "in" doesn't work properly with nested tuples
def find_element(nestedTuple, targetElement):
    for index, element in enumerate(nestedTuple):
        if element == targetElement:
            return True
        if isinstance(element, tuple):
            result = find_element_index(element, targetElement)
            if result:
                return True
    return False
target = 'Coursera'
found = find_element(myTuple, target)

if found:
    print(f"Element {target} is found")
else:
    print(f"Element {target} not found")
```

```python
#Q20 check whether 55 exists within testTuple
testTuple=(1,2,33,55,6,55)
print(55 in testTuple)
```

```python
#Q21 Find the index of JDA in myTuple

def find_element_index(nestedTuple, targetElement, currentIndex=()):
    for index, element in enumerate(nestedTuple):
        if element == targetElement:
            return currentIndex + (index,)
        if isinstance(element, tuple):
            result = find_element_index(element, targetElement, currentIndex + (index,))
            if result is not None:
                return result
    return None
target = 'JDA'
targetIndex = find_element_index(myTuple, target)

if targetIndex is not None:
    print(f"Element {target} found at index {targetIndex}")
else:
    print(f"Element {target} not found in the nested tuple.")
```
